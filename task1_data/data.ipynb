```python
import pandas as pd
```

```python
df=pd.read_csv('rating.csv')
```

```python
df
```

```python
from urllib.parse import urlparse

def extract_domain(url):
    domain = urlparse(url).netloc
    return domain
df['domain'] = df['url'].apply(extract_domain)
domain_counts = df['domain'].value_counts()
print(domain_counts)
```

```python
website_counts = df['domain'].value_counts()
```

```python
top_10_websites = website_counts.head(10)
```

```python
top_10_websites
```

```python
bottom_10_websites = website_counts.tail(10)
```

```python
bottom_10_websites 
```

```python
highest_visitors= df.groupby('domain').size().reset_index(name='article')
```

```python
sort_highest_visitors=highest_visitors.sort_values(by='article', ascending=False)
```

```python
sort_highest_visitors.head(10)
```

```python
sort_highest_visitors.tail(10)
```

```python
country_domain_counts = df.groupby('source_name')['domain'].nunique().reset_index(name='Media Organization Count')
```

```python
sorted_counts = country_domain_counts.sort_values(by='Media Organization Count', ascending=False)
```

```python
sorted_counts.head(10)
```

```python
sorted_counts.tail(10)
```
